```python
import pandas as pd
import numpy as np

# 1. Carga el CSV en df
df = pd.read_csv('../data/muestras_calidad_agua.csv')

# 2. Extrae sitio y temperatura en df_temp
df_temp = df[['sitio', 'temperatura (°C)']].copy()

# Ver resultado
df_temp
```

```python
# 3. Crea categoria_temp con np.where()
# <25 → ÓPTIMA | 25-28 → ALERTA | >28 → CRÍTICA
df_temp['categoria_temp'] = np.where(
    df_temp['temperatura (°C)'] < 25, 'ÓPTIMA',
    np.where(df_temp['temperatura (°C)'] <= 28, 'ALERTA', 'CRÍTICA')
)

df_temp
```

```python
# 4. Crea riesgo_termico: SI / NO
df_temp['riesgo_termico'] = np.where(
    df_temp['temperatura (°C)'] > 28, 'SI', 'NO'
)

df_temp
```

```python
# 5. Filtra a df_caliente (alta temperatura)
df_caliente = df_temp[df_temp['riesgo_termico'] == 'SI']

print(" Sitios con ALTA TEMPERATURA / RIESGO:")
df_caliente
```

```python
# 6. Exporta resultado_temperatura.csv
df_temp.to_csv('resultado_temperatura.csv', index=False)
print(" Archivo guardado: resultado_temperatura.csv")
```

```python
# Reto adicional: porcentaje de sitios con riesgo térmico
total = len(df_temp)
con_riesgo = len(df_caliente)
porcentaje = (con_riesgo / total) * 100

print(f" Porcentaje con RIESGO TÉRMICO: {porcentaje.round(1)}%")
```
