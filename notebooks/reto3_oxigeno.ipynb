# 🧩 Reto 3 – Oxígeno disuelto
**Objetivo:** evaluar riesgo ecológico según oxígeno disuelto.

**Datos:** `../data/muestras_calidad_agua.csv`

### Pasos
1. Carga el CSV en `df`.
2. Extrae `sitio` y `oxigeno`.
3. Crea `nivel_oxigeno` (bajo/medio/alto).
4. Crea `riesgo_biologico` (⚠ si oxígeno < 5).
5. Filtra y exporta `resultado_oxigeno.csv`.

> **Reto adicional:** porcentaje de sitios con riesgo.

```python
# Abrir archivo CSV
with open('../data/muestras_calidad_agua.csv', 'r') as archivo:
    lineas = archivo.readlines()

print("Resultados del análisis de oxígeno:\n")

# Recorrer datos (saltando encabezado)
for linea in lineas[1:]:
    datos = linea.strip().split(',')

    sitio = datos[0]
    oxigeno = float(datos[2])  # Columna de oxígeno

    # Condicionales
    if oxigeno < 5:
        print(sitio, "-", oxigeno, "🔴 ALERTA: Estrés biológico")
    elif oxigeno < 7:
        print(sitio, "-", oxigeno, "🟡 Condición moderada")
    else:
        print(sitio, "-", oxigeno, "🟢 Buen estado ecológico")
```
