En este notebook practicamos condicionales (if, elif, else) sobre *listas, tuplas, diccionarios* y *DataFrames*, y vemos comparaciones vectorizadas con .loc[] y np.where().

```python
# ==============================================
## 1️⃣ Introducción teórica
# ==============================================
print("=== 1. INTRODUCCIÓN TEÓRICA ===")

temperatura = 28
if temperatura > 30:
    print('Día caluroso')
elif temperatura >= 20:
    print('Día templado')
else:
    print('Día frío')


# ==============================================
## 2️⃣ Condicionales con listas
# Selección de un elemento y evaluación
# ==============================================
print("\n=== 2. CONDICIONALES CON LISTAS ===")

ph_list = [6.1, 7.4, 8.9]
ph = ph_list[1]

if ph < 6.5:
    print('El agua es ácida')
elif ph <= 8.5:
    print('El agua es neutra')
else:
    print('El agua es básica')


# ==============================================
## 3️⃣ Condicionales con tuplas
# ==============================================
print("\n=== 3. CONDICIONALES CON TUPLAS ===")

muestra = (29, 4.2)
temperatura = muestra[0]
oxigeno = muestra[1]

if temperatura > 30 and oxigeno < 5:
    print('Alta temperatura y bajo oxígeno → condiciones críticas')
elif temperatura > 30:
    print('Alta temperatura')
elif oxigeno < 5:
    print('Bajo oxígeno')
else:
    print('Condiciones adecuadas')


# ==============================================
## 4️⃣ Condicionales con diccionarios
# ==============================================
print("\n=== 4. CONDICIONALES CON DICCIONARIOS ===")

muestra = {'sitio':'Río Sur','pH':6.3,'oxigeno':5.8}

# Clasificación por pH
if muestra['pH'] < 6.5:
    print(f"{muestra['sitio']}: Agua ácida")
elif muestra['pH'] <= 8.5:
    print(f"{muestra['sitio']}: Agua neutra")
else:
    print(f"{muestra['sitio']}: Agua básica")

# Evaluación de oxígeno
if muestra['oxigeno'] < 5:
    print('Oxígeno bajo → riesgo para organismos acuáticos')
else:
    print('Oxígeno dentro del rango adecuado')
```

```python
# ==============================================
## 5️⃣ Condicionales con DataFrame (una sola muestra)
# ==============================================
print("\n=== 5. CONDICIONALES CON DATAFRAME ===")

import pandas as pd

muestra = pd.DataFrame({'sitio':['Quebrada Verde'],'pH':[8.7],'oxigeno':[4.5]})
valor_ph = muestra.loc[0,'pH']

if valor_ph < 6.5:
    clasificacion = 'ácido'
elif valor_ph <= 8.5:
    clasificacion = 'neutro'
else:
    clasificacion = 'básico'

print(f"Sitio {muestra.loc[0,'sitio']}: agua {clasificacion}")


# ==============================================
## 6️⃣ Condicionales vectorizados en DataFrames
# ==============================================
print("\n=== 6. CONDICIONALES VECTORIZADOS ===")

import numpy as np
import pandas as pd

datos = pd.DataFrame({'sitio':['A','B','C'],'pH':[6.2,7.5,9.1]})
datos['clasificacion'] = np.where(datos['pH'] < 6.5, 'ácido',
                          np.where(datos['pH'] <= 8.5, 'neutro', 'básico'))

datos


# ==============================================
## 7️⃣ Nota complementaria: uso de np.where()
# ==============================================
print("\n=== 7. NOTA COMPLEMENTARIA ===")

import numpy as np

temperaturas = np.array([18,25,32])
resultado = np.where(temperaturas > 30, 'Alta', 'Normal')

print("Temperaturas:", temperaturas.tolist())
print("Clasificación:", resultado.tolist())
```

```python
# ==============================================
## 8  Mini reto RESUELTO
# Carga data/muestras_calidad_agua.csv y crea:
# 1) clasificacion_ph: ácido / neutro / básico
# 2) oxigeno_estado: bajo / medio / alto
# ==============================================
print("\n=== 8.  MINI RETO FINAL ===")

import pandas as pd
import numpy as np

#  Cargamos los datos reales del proyecto
datos = pd.read_csv('../data/muestras_calidad_agua.csv')

#  1) Clasificación por pH
datos['clasificacion_ph'] = np.where(
    datos['pH'] < 6.5, 'ácido',
    np.where(datos['pH'] <= 8.5, 'neutro', 'básico')
)

#  2) Estado del oxígeno
datos['oxigeno_estado'] = np.where(
    datos['oxígeno (mg/L)'] < 5, 'bajo',
    np.where(datos['oxígeno (mg/L)'] <= 7, 'medio', 'alto')
)

#  Resultado final
datos[['sitio', 'pH', 'clasificacion_ph', 'oxígeno (mg/L)', 'oxigeno_estado']]
```
