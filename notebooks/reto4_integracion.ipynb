```python
# =========================
# Función de clasificación
# =========================
def clasificar_biomasa(b):
    if b > 250:
        return 'alta'
    elif 150 <= b <= 250:
        return 'media'
    else:
        return 'baja'

# =========================
# Crear nueva columna
# =========================
df['nivel_biomasa'] = df['biomasa'].apply(clasificar_biomasa)

# Verificar
df.head()

# =========================
# Promedios por nivel
# =========================
promedios = df.groupby('nivel_biomasa')[['temperatura', 'humedad', 'luz']].mean()

print("Promedios por nivel de biomasa:")
print(promedios)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Boxplot temperatura
sns.boxplot(x='nivel_biomasa', y='temperatura', data=df)
plt.title('Temperatura por nivel de biomasa')
plt.show()

# Boxplot humedad
sns.boxplot(x='nivel_biomasa', y='humedad', data=df)
plt.title('Humedad por nivel de biomasa')
plt.show()

# Boxplot luz
sns.boxplot(x='nivel_biomasa', y='luz', data=df)
plt.title('Luz por nivel de biomasa')
plt.show()

# Gráfico de barras de promedios
promedios.plot(kind='bar')
plt.title('Promedios de variables ambientales')
plt.xticks(rotation=0)
plt.show()
```

Se observa que los ecosistemas con biomasa alta presentan mayores valores promedio en variables ambientales como humedad y luz, lo que sugiere condiciones más favorables para la productividad.

Los niveles bajos de biomasa están asociados a condiciones menos óptimas, especialmente en disponibilidad de agua o radiación.

Esto evidencia una relación directa entre las variables ambientales y la producción de biomasa en los ecosistemas analizados.
